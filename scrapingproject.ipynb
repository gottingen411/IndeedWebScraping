```python
#process company
df['company']=df['company'].str.lower()
df['company'] = df['company'].apply(lambda x: re.sub('\s+', ' ', x))
```

```python
# create list of skills to look up 
skill_list=['python','scala',' r ','java','sql','spark','scikit-learn','tensorflow','torch',' aws ','azure',
            'gcp','redhat','nosql','cassandra','elasticsearch','kafka','docker','kubernetes', 'c\+\+', 
            ' cuda ','java','python','mysql','postgresql','hadoop','hbase','hive','redshift','dask','airflow',
            'luigi','svm','kernel functions','gradient descent','back propagation','tableau','looker','kibana',
            'sparksql','sap','teradata','jmp','sap','spss','perl','php','linear','logistic','svm','random forest',
            'k-means clustering', 'k-nearest neighbors','knn',' pca ','bayesian','cognos','ec2','rds','re:dash',
            'superset','ruby','s3',' emr ','graph-based','sparkml','caret','mlr','mllib',' sas ', 'multivariate', 
            'stochastic','sampling'] 
```

```python
# delete duplicates in skill_list
skill_listset=set(skill_list)
skill_list=list(skill_listset)

# search for all these skill keywords in job search results 
import numpy as np
len_search=len(skill_list)
skill_search_count=np.zeros(len_search)
text_list=df['text'].tolist()
skill_search_fun=lambda x,txt: re.search(x, txt)

for ind in range(len_search):
    skill_search=[skill_search_fun(skill_list[ind],t) for t in text_list]
    for result in skill_search:
    #search_len_array=list(map(lambda x: len(x),skill_search))
    #skill_sum=int(sum(search_len_array))
        if result:
            skill_search_count[ind]=skill_search_count[ind]+1

search_dict={'skill':skill_list,'skill_count':skill_search_count}

import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"
search_df=pd.DataFrame(data=search_dict)
search_df=search_df.sort_values(by=['skill_count'],ascending=False)
search_df['skill_count']=search_df['skill_count'].astype(int)
top=search_df.head(20)


plt.figure();

skillbar=top.plot.bar(x='skill',y='skill_count',figsize=(10,15),legend=False)

skillbar.set_ylabel('# job postings',fontsize=20)
skillbar.set_xlabel('',fontsize=20)
for tick in skillbar.get_xticklabels():
    tick.set_fontsize(14)
for tick in skillbar.get_yticklabels():
    tick.set_fontsize(14)
plt.yticks(rotation=90)
plt.savefig('midlevel_skillbar_top.png')
```

```python
# plot time series for each top company
company_datedf=df[df.company.isin(companies)].reset_index()
ldf=len(company_datedf.title)
lcom=len(companies)
comarray=np.asarray(companies)
company_datedf['comid']=0
for c in companies:
    company_datedf.loc[company_datedf.company==c,'comid']=np.where(comarray==c)
company_datedf   
dates=company_datedf.date.sort_values().unique()
ldates=len(dates)


comdate_mat = np.zeros((ldates,lcom+1)).astype(int)

comdate_mat[:,0]=dates
for ind,date in enumerate(dates):
    for ind1 in range(ldf): 
        if company_datedf.loc[ind1,'date']==date:
            temp1=company_datedf['comid'][ind1]
            comdate_mat[ind,temp1+1]=comdate_mat[ind,temp1+1]+1

companyCols=['days']+companies
company_datedfnew=pd.DataFrame(data=comdate_mat,columns=companyCols)
plt.figure()
ax=company_datedfnew.plot(x=0,figsize=(15,10))
ax.set_ylabel('# postings',fontsize=20)
ax.set_xlabel('# days ago',fontsize=20)
for tick in ax.get_xticklabels():
    tick.set_fontsize(16)
for tick in ax.get_yticklabels():
    tick.set_fontsize(16)

plt.savefig('midlevel_companytime.png')


```
